```python
# Scatter plot for 'price' vs. 'rating'
plt.figure(figsize=(8, 6))
plt.scatter(df['price'], df['rating'], alpha=0.5)
plt.title('Price vs. Rating')
plt.xlabel('Price')
plt.ylabel('Rating')
plt.show()
```

```python
# Scatter plot for 'num_reviews' vs. 'rating'
plt.figure(figsize=(8, 6))
plt.scatter(df['num_reviews'], df['rating'], alpha=0.5)
plt.title('Number of Reviews vs. Rating')
plt.xlabel('Number of Reviews')
plt.ylabel('Rating')
plt.show()
```

```python
# Scatter plot for 'price' vs. 'rating' using Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='price', y='rating', alpha=0.5)
plt.title('Price vs. Rating')
plt.xlabel('Price')
plt.ylabel('Rating')
plt.show()
```

```python
# Scatter plot for 'num_reviews' vs. 'rating' using Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='num_reviews', y='rating', alpha=0.5)
plt.title('Number of Reviews vs. Rating')
plt.xlabel('Number of Reviews')
plt.ylabel('Rating')
plt.show()
```

```python
#BOx plot for outlayer checking
for col in df:
  plt.figure(figsize=(6,6))
  sns.boxplot(df[col])
  plt.show()
```

```python
g = sns.pairplot(data=df, diag_kind="kde", dropna=True)
g.map_lower(sns.kdeplot, levels=4, color=".2")
```
